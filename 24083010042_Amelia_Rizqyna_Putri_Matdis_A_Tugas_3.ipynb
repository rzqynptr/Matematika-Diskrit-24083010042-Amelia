```python
import matplotlib.pyplot as plt

# Fungsi untuk menghitung jumlah parsial deret harmonik
def harmonic_partial_sum(n):
    return sum(1/i for i in range(1, n+1))

# --- Bagian 1: Cetak nilai sampai melewati 2 ---
for n in range(1, 100):
    s = harmonic_partial_sum(n)
    print(f"n={n}, H_n={s:.6f}")
    if s > 2:
        print(f"\n--> Pertama kali H_n lebih dari 2 terjadi pada n={n}, dengan nilai {s:.6f}\n")
        break

# --- Bagian 2: Buat grafik ---
n_values = list(range(1, 51))
sums = [harmonic_partial_sum(n) for n in n_values]

plt.figure(figsize=(8,5))
plt.plot(n_values, sums, marker="o", color="orange", label="Harmonic Partial Sum")
plt.axhline(y=2, color="red", linestyle="--", label="y=2")

plt.title("Jumlah Parsial Deret Harmonik")
plt.xlabel("n (jumlah suku)")
plt.ylabel("H_n = 1 + 1/2 + ... + 1/n")
plt.legend()
plt.grid(True)
plt.show()
```

Jadi, kode ini pertama menghitung jumlah parsial dari deret harmonik, yaitu 1 + 1/2 + 1/3 .... + 1/n. Program akan menampilkan hasil setiap kali ùëõ bertambah, lalu dicek kapan nilai jumlah ini pertama kali melebihi angka 2. Dari output terlihat bahwa pada ùëõ =4, jumlah parsialnya sudah lebih besar dari 2, yaitu sekitar 2,0833.

Kemudian, di bagian kedua, kode membuat grafik dari jumlah parsial deret harmonik sampai ùëõ = 50. Grafik ini menunjukkan bahwa nilai deretnya terus naik meskipun kenaikannya semakin kecil. Garis merah putus-putus di posisi ùë¶ = 2 dipakai sebagai batas pembanding. Dari grafik bisa kita lihat dengan jelas bahwa sejak ùëõ = 4, garis deret harmonik sudah melampaui angka 2 dan terus naik tanpa berhenti.
