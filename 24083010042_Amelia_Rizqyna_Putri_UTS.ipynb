**Abstract**: One trending theme within research on learning and teaching is an emphasis on artificial intelligence.
```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from wordcloud import WordCloud, STOPWORDS

# -------------------------------
# 1️⃣ BACA TEKS DARI FILE .TXT
# -------------------------------
# Pastikan file .txt kamu ada di folder kerja yang sama
with open("teks artikel matdis.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
# -------------------------------
# 2️⃣ PEMBERSIHAN TEKS
# -------------------------------
clean_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
words = clean_text.split()
```

```python
# -------------------------------
# 3️⃣ HITUNG KONSONAN & VOKAL
# -------------------------------
def count_vowels(word):
    return len(re.findall(r'[aeiou]', word))

def count_consonants(word):
    return len(re.findall(r'[bcdfghjklmnpqrstvwxyz]', word))

df = pd.DataFrame({
    'word': words,
    'vowels': [count_vowels(w) for w in words],
    'consonants': [count_consonants(w) for w in words]
})
```

```python
# -------------------------------
# 4️⃣ TABEL PIVOT & HEATMAPS
# -------------------------------
pivot_freq = pd.pivot_table(df, index='consonants', columns='vowels',
                            values='word', aggfunc='count', fill_value=0)

pivot_unique = df.groupby(['consonants', 'vowels'])['word'].nunique().unstack(fill_value=0)

print("=== TABEL PIVOT: FREKUENSI TOKEN ===")
print(pivot_freq)
print("\n=== TABEL PIVOT: CACAH KATA UNIK ===")
print(pivot_unique)

fig, axes = plt.subplots(1, 2, figsize=(16,6))
sns.heatmap(pivot_freq, annot=True, cmap='YlGnBu', ax=axes[0])
axes[0].set_title("Frekuensi Kata (Token)")
axes[0].set_xlabel("Jumlah Vokal")
axes[0].set_ylabel("Jumlah Konsonan")

sns.heatmap(pivot_unique, annot=True, cmap='YlOrBr', ax=axes[1])
axes[1].set_title("Cacah Kata Unik (Distinct Words)")
axes[1].set_xlabel("Jumlah Vokal")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

Hasil analisis menunjukkan bahwa kata dengan 2-3 vokal dan 2-4 konsonan paling sering muncul dalam teks. Hal ini menandakan bahwa struktur kata dalam artikel ilmiah tersebut cenderung pendek hingga sedang, sehingga mudah dibaca dan sesuai dengan gaya bahasa akademik.

Frekuensi kata tinggi menunjukkan adanya pengulangan istilah penting seperti "AI", "education", dan "learning", sedangkan hasil cacah kata menandakan adanya keragaman kosakata yang tetap terjaga.

```python
# -------------------------------
# 5️⃣ JUMLAH KATA PER KALIMAT
# -------------------------------
sentences = re.split(r'(?<=[.!?]) +', text.strip())
sentence_word_counts = [len(s.split()) for s in sentences]

plt.figure(figsize=(7,5))
plt.hist(sentence_word_counts, bins=10, edgecolor='black')
plt.title("Distribusi Jumlah Kata per Kalimat")
plt.xlabel("Jumlah Kata")
plt.ylabel("Frekuensi")
plt.show()

print("\n=== JUMLAH KATA PER KALIMAT ===")
for i, count in enumerate(sentence_word_counts, 1):
    print(f"Kalimat {i}: {count} kata")
```

Berdasarkan hasil histogram, sebagian besar kalimat dalam teks memiliki 15-30 kata, dengan puncak frekuensi berada pada kisaran 20 kata per kalimat. Hal ini menunjukkan bahwa struktur kalimat dalam artikel cenderung pendek hingga sedang, yang memudahkan pembaca dalam memahami isi teks.

Beberapa kalimat memiliki jumlah kata lebih panjang (sekitar 60-90 kata), yang kemungkinan merupakan kalimat kompleks atau deskriiptif khas penulisan ilmiah.


```python
# -------------------------------
# 6️⃣ JUMLAH KALIMAT PER PARAGRAF
# -------------------------------
paragraphs = [p for p in text.split("\n\n") if p.strip()]
paragraph_sentence_counts = [len(re.split(r'(?<=[.!?]) +', p.strip())) for p in paragraphs]

plt.figure(figsize=(6,4))
plt.bar(range(1, len(paragraph_sentence_counts)+1), paragraph_sentence_counts)
plt.title("Jumlah Kalimat per Paragraf")
plt.xlabel("Paragraf ke-")
plt.ylabel("Jumlah Kalimat")
plt.show()

print("\n=== JUMLAH KALIMAT PER PARAGRAF ===")
for i, count in enumerate(paragraph_sentence_counts, 1):
    print(f"Paragraf {i}: {count} kalimat")
```

Berdasarkan hasil diagram batang, jumlah kalimat pada setiap bagian artikel menunjukkan bahwa abstrak terdiri dari 10 kalimat, sedangkan pendahuluan dan kesimpulan masing-masing memiliki 11 kalimat.

Temuan ini menunjukkan bahwa setiap bagian artikel memiliki panjang yang relatif seimbang, dengan hanya sedikit perbedaan jumlah kalimat antarbagian. Struktur seperti ini mencerminkan konsistensi penulisan akademik, di mana setiap bagian memiliki porsi informasi yang proporsional. Selain itu, keseimbangan jumlah kalimat juga dapat memudahkan pembaca dalam memahami alur penjelasan dari bagian awal hingga penutup artikel.

```python
# -------------------------------
# 7️⃣ WORDCLOUD
# -------------------------------
stopwords = set(STOPWORDS)
text_for_cloud = " ".join(words)

wordcloud = WordCloud(
    width=1000, height=600,
    background_color='white',
    stopwords=stopwords,
    max_words=100,
    colormap='viridis',
    contour_color='steelblue',
    contour_width=1
).generate(text_for_cloud)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud Artikel AI", fontsize=16)
plt.show()
```

Berdasarkan hasil word cloud artikel bertema AI dalam pendidikan, kata-kata yang paling menonjol adalah “learning”, “education”, “research”, “teaching”, “human”, dan “AI”. Dominasi kata-kata tersebut menunjukkan bahwa fokus utama artikel berkaitan dengan penerapan kecerdasan buatan (AI) dalam proses pembelajaran dan penelitian pendidikan.

Selain itu, munculnya istilah seperti “data”, “process”, “ethical”, dan “challenges” menandakan adanya pembahasan tentang tantangan etika, pengelolaan data, serta potensi AI dalam meningkatkan efektivitas pendidikan.
